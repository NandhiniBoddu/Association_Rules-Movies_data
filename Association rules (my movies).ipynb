# Apriori Algorithm

```python
frequent_itemsets= apriori(data, min_support=0.1, use_colnames=True)
frequent_itemsets 
```

```python
frequent_itemsets= apriori(data1, min_support=0.1, use_colnames=True)
frequent_itemsets 
```

```python
frequent_itemsets2= apriori(data1, min_support=0.2, use_colnames=True)
frequent_itemsets2 
```

```python
rules= association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)
rules 
```

```python
rules1= association_rules(frequent_itemsets, metric="lift", min_threshold=5)
rules1
```
